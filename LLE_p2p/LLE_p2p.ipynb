```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import manifold, datasets
```

```python
data = pd.read_csv('./p2p.csv', index_col=0)
```

```python
y = data.status
data = data.drop(['status', 'nace'], axis=1)
```

```python
X_r, err = manifold.locally_linear_embedding(data, n_neighbors=12,
                                             n_components=2)
```

```python
# LLE

# Plot data with dimensionality reduction PCA
fig = plt.figure(figsize=(10, 10))
plt.scatter(X_r[:,0], X_r[:,1], c=y, marker='o', cmap=plt.cm.coolwarm)

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('Reduced to 2D using LLE')

plt.savefig('./LLE_reduce.png', transparent=True)
```
